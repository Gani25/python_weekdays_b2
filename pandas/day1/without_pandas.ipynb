# Find Max Rating and Min Rating and Average Rating Of All Movies

```python
imdb_ratings = []

for row in data:
    if(row[3] != "NULL"):
        imdb_ratings.append(float(row[3]))

print(imdb_ratings)
```

```python
max_rating_all_movies = max(imdb_ratings)
min_rating_all_movies = min(imdb_ratings)
avg_rating_all_movies = sum(imdb_ratings) / len(imdb_ratings)

print("All Movies Ratings Info")
print(f"Max Rating = {max_rating_all_movies}")
print(f"Min Rating = {min_rating_all_movies}")
print(f"Average Rating = {round(avg_rating_all_movies,2)}")
```
